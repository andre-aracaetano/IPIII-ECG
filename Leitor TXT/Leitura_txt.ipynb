```python
todos_arquivos = []
b = 0

for i in range(30):
    f = 0
    n = f'ANDRE{i + 1}.txt'
    with open(n, 'r') as file:
        for linha in file:
            linha = linha.strip()
            todos_arquivos.append(linha)
            f = 1 + f
    b = f/60 + b
    todos_arquivos = [int(x) for x in todos_arquivos]
    print("Tarefa concluda no arquivo:", n)

todos_arquivos = [x / 1000 for x in todos_arquivos]
b = b/30
print(b)
```

```python
time1 = []
sinal1 = []

with open('a.txt', 'r') as file:
    for linha in file:
        linha = linha.strip()
        valores = linha.split(';')
        time1.append(valores[0])
        sinal1.append(valores[1])

time1.remove("Tempo (ms)")
sinal1.remove("Valor")

time1 = [int(x) for x in time1]
sinal1 = [int(x) for x in sinal1]

#time1 = [x / 1000 for x in time1]

print("Coluna 1:", time1)
print("Coluna 2:", sinal1)
```
